##Â Feature Extraction

The code below extract the audio feature by segmenting each audio file in small chunks, with a duration of 20ms. After that, a Complex-Q Transformation will be applied to the chunk audio data, resulting in an image containing the frequency information.
The annotated information related to the same audio chunk, the ground truth, is also extracted.

The result of the data processing step is:

x_train_opt array, containing an array of over 50000 images taken from the Complex-Q Transformation
y_train_one_hot array, consisting of 6 arrays of the same x_train_opt length, containing annotations for every audio chunk for each string

```python
import librosa as _librosa
import librosa.display as _display
from presets import Preset

import numpy as np
import matplotlib.pyplot as plt

from os import listdir, mkdir, makedirs, rmdir
from os.path import isfile, join, isdir
from shutil import rmtree

import pickle

import pandas as pd

import jams

librosa = Preset(_librosa)
librosa.display = _display

duration= 0.2
strings = [ 'e string', 'A string', 'D string', 'G string', 'B string', 'E string']

notesInString = [['e string OFF', 'E2', 'F2', 'F#2', 'G2', 'G#2', 'A2', 'A#2', 'B2', 
                  'C3', 'C#3', 'D3', 'D#3', 'E3', 'F3', 'F#3', 'G3', 'G#3', 'A3', 'A#3', 'B3'],
                 ['A string OFF', 'A2', 'A#2', 'B2', 'C3', 'C#3', 'D3', 'D#3', 'E3', 
                  'F3', 'F#3', 'G3', 'G#3', 'A3', 'A#3', 'B3', 'C4', 'C#4', 'D4', 'D#4', 'E4'],
                 ['D string OFF', 'D3', 'D#3', 'E3', 'F3', 'F#3', 'G3', 'G#3', 'A3', 
                  'A#3', 'B3', 'C4', 'C#4', 'D4', 'D#4', 'E4', 'F4', 'F#4', 'G4', 'G#4', 'A4'],
                 ['G string OFF', 'G3', 'G#3', 'A3', 'A#3', 'B3', 'C4', 'C#4', 'D4', 
                  'D#4', 'E4', 'F4', 'F#4', 'G4', 'G#4', 'A4', 'A#4', 'B4', 'C5', 'C#5', 'D5'],
                 ['B string OFF', 'B3', 'C4', 'C#4', 'D4', 'D#4', 'E4', 'F4', 'F#4', 
                  'G4', 'G#4', 'A4', 'A#4', 'B4', 'C5', 'C#5', 'D5', 'D#5', 'E5', 'F5', 'F#5'],
                 ['E string OFF', 'E4', 'F4', 'F#4', 'G4', 'G#4', 'A4', 'A#4', 'B4', 
                  'C5', 'C#5', 'D5', 'D#5', 'E5', 'F5', 'F#5', 'G5', 'G#5', 'A5', 'A#5', 'B5']]
```

```python
dataFolder = 'data/GuitarSet/audio/audio_mic'
annotationFolder = 'data/GuitarSet/annotation'
x_train = []
y_train = [[] for i in range (6)]

def getNoteClassification(note, string):    
    if note in notesInString[string]:
        return notesInString[string].index(note)
    raise ValueError('Exception Note', note, 'not in string', string) 

print('Recreating training data...')
dirs = listdir(dataFolder)
dirs.sort()
for f in dirs:
    if '.DS_Store' not in f:
        print('Processing file', f)
        y, sr = librosa.load(join(dataFolder, f))
        curTime = 0
        offset = int(sr*curTime)
        chunk = int(sr*duration)
        annotationFile = f.replace('_mic.wav', '.jams')

        with open(join(annotationFolder, annotationFile)) as jams_file:
            data = jams.load(jams_file)

            observations = [[] for i in range (6)]
            indexes = [0 for i in range(6)]
            for a in data['annotations']:            
                if a['namespace'] == 'pitch_contour':
                    currentIndex = int(a['annotation_metadata']['data_source'])
                    observations[currentIndex] = a['data']

            while offset + chunk <= len(y):
                # Retrieves the annotations for the audio segment (curTime)
                chunk = int(sr*duration)        
                lastIndex = len(y_train[0])
                try:
                    for i in range(len(indexes)):                    
                        obs = observations[i]
                        idx = indexes[i]
                        if len(obs) == 0:
                            y_train[i].append(getNoteClassification(strings[i] + ' OFF', i))
                        else:
                            while obs[idx].time < curTime:
                                if idx < len(obs)-1:
                                    idx += 1
                                else:
                                    break
                            indexes[i] = idx

                            if obs[idx].time <= curTime+duration:
                                freq = obs[idx].value['frequency']
                                if int(freq) == 0:
                                    y_train[i].append(getNoteClassification(strings[i] + ' OFF', i))
                                else:
                                    y_train[i].append(getNoteClassification(librosa.hz_to_note(freq), i))
                            else:
                                y_train[i].append(getNoteClassification(strings[i] + ' OFF', i))

                    # Creates a CQT for the specified audio segment 
                    x_train.append(np.abs(librosa.cqt(y[offset:offset+chunk], sr=sr, 
                                           n_bins=7*12, fmin=librosa.note_to_hz('C2'),
                                           bins_per_octave=12)))


                except Exception as e:
                    # If we detect a wrong annotation, such as a note which is the result
                    # of a resonance rather then the string being strummed
                    # then we discard any annotations related to this time interval
                    for i in range(len(indexes)):
                        if len(y_train[i]) > lastIndex:
                            y_train[i].pop()

                curTime += 0.2
                offset += chunk

print('Done, dumping x_train data...')
pickle.dump(np.array(x_train), open('data/x_train.data', 'wb'))
print('Done, dumping y_train data...')                
pickle.dump(y_train, open('data/y_train.data', 'wb'))

print('Data dumped.')
```
